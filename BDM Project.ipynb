```python
import pandas as pd
from datetime import datetime

# Load the data
df = pd.read_csv(r'C:\\Users\\kabali\\OneDrive\\Desktop\\OPTICALS_Cleaned.csv')

# Convert the date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Extract the month
df['Month'] = df['Date'].dt.month

# Extract the relevant features
df = df[['Month', 'Brand', 'Type', 'Shape', 'Quantity', 'Total_Price']]

# Group the data by month and brand
df_grouped = df.groupby(['Month', 'Brand']).agg({'Quantity': 'sum', 'Total_Price': 'mean'}).reset_index()

# Display the preprocessed data
print(df_grouped.head())
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
import warnings
warnings.warn("ignore")
# Create a list of unique brands
brands = df_grouped['Brand'].unique()

# Create an empty dataframe to store the predictions
predictions = pd.DataFrame(columns=['Month', 'Brand', 'Quantity', 'Total_Price'])

# Loop through each brand
for brand in brands:
    # Filter the data for the brand
    data = df_grouped[df_grouped['Brand'] == brand]
    
    # Extract the features and target
    X = data[['Month']]
    y_quantity = data['Quantity']
    y_price = data['Total_Price']
    
    # Fit a linear regression model for quantity
    model_quantity = DecisionTreeRegressor()
    model_quantity.fit(X, y_quantity)
    
    # Predict the quantity for the next month
    next_month_quantity = model_quantity.predict([[df_grouped['Month'].max() + 1]])[0]
    
    # Fit a linear regression model for price
    model_price = LinearRegression()
    model_price.fit(X, y_price)
    
    # Predict the price for the next month
    next_month_price = model_price.predict([[df_grouped['Month'].max() + 1]])[0]
    
    # Add the predictions to the dataframe
    predictions = predictions.append({
        'Month': df_grouped['Month'].max() + 1,
        'Brand': brand,
        'Quantity': next_month_quantity,
        'Total_Price': next_month_price
    }, ignore_index=True)

# Display the predictions
print(predictions.head())
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate MAPE
def mape(actual, predicted):
    return np.mean(np.abs((actual - predicted) / actual)) * 100

# Calculate RMSE
def rmse(actual, predicted):
    return np.sqrt(mean_squared_error(actual, predicted))
```
